### Identify Duplicate Rows

### Group by State & Descriptive Statistics

#### (Mean, Median, Min, Max for numeric columns)

```python
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns.drop('Business ID')
numeric_columns
```

```python
state_stats = df.groupby("Business State")[numeric_columns].agg(
    ['mean', 'median', 'min', 'max']
)
```

```python
state_stats.head()
```

### Filter Businesses with Negative Debt-to-Equity Ratio

```python
df["Debt to Income"] = np.where(
    df["Total Revenue"] == 0,
    np.nan,
    df["Total Long-term Debt"] / df["Total Revenue"]
)
df.head()
```

```python
print("Businesses with negative Debt-to-Equity ratio:", df["Debt to Income"].shape[0])
```

### Create Debt-to-Income Ratio DataFrame

```python
debt_to_income_df = pd.DataFrame()
```

```python
debt_to_income_df = pd.DataFrame()

debt_to_income_df["Debt_to_Income_Ratio"] = np.where(
    df["Total Revenue"] == 0,
    np.nan,
    df["Total Long-term Debt"] / df["Total Revenue"]
)
```

```python
debt_to_income_df.head()
```

### Concatenate with Original DataFrame

```python
final_df = pd.concat([df, debt_to_income_df], axis=1)
```

```python
final_df.head()
```
